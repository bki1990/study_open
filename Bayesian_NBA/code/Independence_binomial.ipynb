```r
m <- 5000 #length of chain
xt <- numeric(m)
a <- 1 #parameter of Beta(a,b) proposal dist.
b <- 1 #parameter of Beta(a,b) proposal dist.
n <- 4 #sample size
```

```r

# generate the observed sample
x <- c(1,1,1,0)
```

```r
# generate the independence sampler chain
u <- runif(m)
y <- rbeta(m, a, b) #proposal distribution
```

```r
xt[1] <- .5
```

```r
for (i in 2:m) {
    fy <- (y[i]^3)*(1-y[i])
    fx <- (xt[i-1]^3)*(1-xt[i-1])

    r <- (fy / fx) *(xt[i-1]^(a-1) * (1-xt[i-1])^(b-1)) /(y[i]^(a-1) * (1-y[i])^(b-1))

    if (u[i] <= r) {xt[i] <- y[i]}
    else {
    xt[i] <- xt[i-1]}
}
```

```r
plot(xt, type="l", ylab="p")
hist(xt[((m/2) +1):m], main="", xlab="p", prob=TRUE)
print(mean(xt[101:m]))
```
